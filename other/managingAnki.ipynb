```python
from anki.collection import Collection
import re
import pandas as pd

# Replace this with the path to your Anki profile folder
profile_folder = "/Users/moritzvitt/Library/Application Support/Anki2/moritzelias.vitt@gmail.com"
```

```python
try:
    # Open the Anki collection for the specific profile
    col = Collection(profile_folder + "/collection.anki2")

    # Get a list of all decks in the collection
    all_decks = col.decks.all()

    if all_decks:
        # Create a set to store subdeck names
        subdeck_names = set()

        # Iterate through the list of decks
        for deck_info in all_decks:
            deck_name = deck_info['name']
            # Check if this deck's name is in any other deck's name (subdeck)
            for other_deck_info in all_decks:
                other_deck_name = other_deck_info['name']
                if deck_name in other_deck_name and deck_name != other_deck_name:
                    subdeck_names.add(deck_name)

        # Print only top-level decks
        for deck_info in all_decks:
            deck_name = deck_info['name']
            if deck_name not in subdeck_names:
                print(f"Top-Level Deck Name: {deck_name}")

    else:
        print("No decks found in the collection.")

except Exception as e:
    print("Error:", e)
```
